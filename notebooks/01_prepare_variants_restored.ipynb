```python
from pathlib import Path
import pandas as pd

project_root = Path.cwd()

if project_root.name == "notebooks":
    project_root = project_root.parent

project_root
```

```python
csv_file = (
    project_root
    / "data"
    / "processed"
    / "atm_vus_app_dataset.csv"
)

df = pd.read_csv(csv_file)
df.head()
```

```python
#filter to only include parsed and remove duplicates
before_filter = len(df)
prepared_df = df[df["protein_parse_status"] == "Parsed"].drop_duplicates(subset=["clinvar_variation_id"]).copy()

after_filter = len(prepared_df)
removed =  before_filter - after_filter

print(f"{removed} rows filtered out; {after_filter} of {before_filter} rows remain.")
```

```python
# all are cleaned already
print(df["protein_parse_status"].value_counts(dropna=False))
print("Duplicate ClinVar IDs:", df["clinvar_variation_id"].duplicated().sum())
```

```python
output_file = (
    project_root
    / "data"
    / "processed"
    / "prepared_atm_missense_vus.csv"
)

prepared_df.to_csv(output_file)
print(f"Saved to: {output_file}")
```
